```python
''' Create the folder structure/directories and files using the python programme required for a ML project.
You can refer the following project structure
< sr3
, __init__.p*
, logger.p*
, exception.p*
, utils.p*
, component5
, __init__.p*
, data_ingestion.p*
, data_transformation.p*
, model_trainer.p*
, pipelin4
, __init__.p*
, predict_pipeline.p*
, train_pipeline.p*
< import_data.p*
< setup.p*
< noteboo+
< requirements.txt'''


import os

# Define the project structure
project_structure = {
    "sr3": [
        "__init__.py",
        "logger.py",
        "exception.py",
        "utils.py",
        {
            "components": [
                "__init__.py",
                "data_ingestion.py",
                "data_transformation.py",
                "model_trainer.py",
            ]
        },
        {
            "pipeline": [
                "__init__.py",
                "predict_pipeline.py",
                "train_pipeline.py",
            ]
        }
    ],
    "import_data.py": None,
    "setup.py": None,
    "notebooks": None,
    "requirements.txt": None
}

def create_project_structure(base_path, structure):
    """
    Recursively create directories and files based on the provided structure.
    """
    for key, value in structure.items():
        path = os.path.join(base_path, key)

        if isinstance(value, list):
            # Create a directory and files within it
            os.makedirs(path, exist_ok=True)
            for item in value:
                if isinstance(item, dict):
                    # Recursive call to create subdirectories
                    create_project_structure(path, item)
                else:
                    # Create files inside the directory
                    file_path = os.path.join(path, item)
                    open(file_path, 'w').close()  # Create an empty file
        else:
            # Create a single file at the base path
            open(path, 'w').close()

# Set the base path to your desired directory
base_path = "ml_project"

# Create the entire project structure
create_project_structure(base_path, project_structure)

print(f"Project structure created in {base_path}")


!python create_project_structure.py

!cd ml_project
!git init
!git add .

!git commit -m "Initial commit of ML project structure"
!git remote add origin https://github.com/your-username/your-repo.git
!git push -u origin master


'''Pull the Changes to Your Local Repository:'''

!git pull origin master
# Machine Learning Project

'''This project contains code and components for a machine learning pipeline including data ingestion, transformation, and model training.

## Folder Structure
- sr3: Main project folder containing modules like logging, exceptions, and utilities.
- components: Contains the scripts for data ingestion, data transformation, and model training.
- pipeline: Contains pipelines for training and prediction.

## How to Run
Instructions on running the project go here.'''

# Ignore virtual environment folders
"venv/"
".env/"

# Ignore Python cache files
"__pycache__/"
"*.pyc"
!git config --global user.email "kingshukroy35@gmail.com"
!git config --global user.name "kingshuk-roy"
```

```python
'''Write the logging function in logger.py and exception function in exception.py file to be used for the
project to track the progress when the ML project is run and to raise any exception when encountered.'''


import logging
import os
from datetime import datetime

# Create a logs directory if it doesn't exist
LOG_DIR = "logs"
os.makedirs(LOG_DIR, exist_ok=True)

# Generate a unique log file name with a timestamp
LOG_FILE = f"{LOG_DIR}/ml_project_{datetime.now().strftime('%Y_%m_%d_%H_%M_%S')}.log"

# Configure logging
logging.basicConfig(
    filename=LOG_FILE,  # Log file path
    filemode='a',  # Append mode
    format='%(asctime)s - %(levelname)s - %(message)s',  # Log format
    level=logging.INFO  # Log level (INFO, DEBUG, ERROR, etc.)
)

# Create a logger object
logger = logging.getLogger()

def log_info(message):
    """
    Logs an INFO message.
    """
    logger.info(message)

def log_error(message):
    """
    Logs an ERROR message.
    """
    logger.error(message)

def log_debug(message):
    """
    Logs a DEBUG message.
    """
    logger.debug(message)

def log_warning(message):
    """
    Logs a WARNING message.
    """
    logger.warning(message)

'''Custom Exception Handling'''


import sys
import logging

# exception.py
import sys
import logging

# Change from relative to absolute import
#from .logger import log_error
from logger import log_error

# ... (rest of your exception.py code)


#from .logger import log_error
class CustomException(Exception):
    def __init__(self, message, error_detail: sys):
        """
        Custom Exception class to log error messages with detailed information.

        Args:
            message (str): Error message to display.
            error_detail (sys): System information for additional error context.
        """
        super().__init__(message)
        self.error_detail = error_detail

    def __str__(self):
        """
        Custom string representation to display a detailed error message.

        Returns:
            str: Formatted error message with traceback details.
        """
        _, _, exc_tb = self.error_detail.exc_info()
        # Extract file name and line number from the traceback
        file_name = exc_tb.tb_frame.f_code.co_filename
        line_number = exc_tb.tb_lineno
        return f"Error occurred in script: [{file_name}] at line number: [{line_number}] | Message: {str(self)}"

def log_and_raise_exception(message):
    """
    Logs the error and raises a custom exception.

    Args:
        message (str): Error message.
    """
    exc_type, exc_obj, exc_tb = sys.exc_info()
    if exc_type:
        error_message = f"Exception Type: {exc_type.__name__}, Error: {str(exc_obj)}, Line: {exc_tb.tb_lineno}"
        log_error(error_message)
    raise CustomException(message, sys)
```
